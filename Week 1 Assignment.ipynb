## Step 1

```python
# Load the packages

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data. 
# EPL results for the 2017/18 season

EPL18 = pd.read_excel('EPL2017-18.xlsx')
print(EPL18.columns.tolist())
```

## Step 2

```python
EPL18['home_win'] = np.where(EPL18['FTHG'] > EPL18['FTAG'], 1, 
                             np.where(EPL18['FTHG'] == EPL18['FTAG'], 0.5, 0))
EPL18['away_win'] = np.where(EPL18['FTHG'] < EPL18['FTAG'], 1, 
                             np.where(EPL18['FTHG'] == EPL18['FTAG'], 0.5, 0))
EPL18['count'] = 1
```

## Step 3

Create a file for games played in 2017 (before date 20180000) and another one for games played in 2018 (after date 20180000). For the 2017 games, use .groupby to create a dataframe aggregating by home team the variables for count, home wins, goals for and goals against. Then, use .groupby to create a separate dataframe aggregating by away team the variables for count, away wins, goals for and goals against. Rename the variables to denote whether they are aggregates for home team or away team.

```python
EPL2017 = EPL18[ EPL18['Date'] < 20180000 ]
EPL2018 = EPL18[ EPL18['Date'] >= 20180000 ]
```

## Step 4 (home team, 2017)

```python
EPL2017_home = EPL2017[['HomeTeam','FTHG','FTAG','home_win']].copy()
EPL2017_home['home_games']=1
EPL2017_home = EPL2017_home.rename(columns={'HomeTeam':'Team','FTHG':'home_R','FTAG':'home_RA','count':'home_games'})
```

```python
EPL2017_home_by_teams = EPL2017_home.groupby('Team')['home_games','home_win','home_R','home_RA'].sum().reset_index()
EPL2017_home_by_teams
```

## Step 5 (home team, 2018)

```python
EPL2018_home = EPL2018[['HomeTeam','FTHG','FTAG','home_win']].copy()
EPL2018_home['home_games']=1
EPL2018_home = EPL2018_home.rename(columns={'HomeTeam':'Team','FTHG':'home_R','FTAG':'home_RA'})
```

```python
EPL2018_home_by_teams = EPL2018_home.groupby('Team')['home_games','home_win','home_R','home_RA'].sum().reset_index()
EPL2018_home_by_teams
```

## Step 7 (=Step 4 (away team, 2017))

```python
EPL2017_away = EPL2017[['AwayTeam','FTHG','FTAG','away_win']].copy()
EPL2017_away['away_games']=1
EPL2017_away = EPL2017_away.rename(columns={'AwayTeam':'Team','FTAG':'away_R','FTHG':'away_RA'})
```

```python
EPL2017_away_by_teams = EPL2017_away.groupby('Team')['away_games','away_win','away_R','away_RA'].sum().reset_index()
EPL2017_away_by_teams
```

## Step 7 (=Step 5 (away team, 2018))

```python
EPL2018_away = EPL2018[['AwayTeam','FTHG','FTAG','away_win']].copy()
EPL2018_away['away_games']=1
EPL2018_away = EPL2018_away.rename(columns={'AwayTeam':'Team','FTAG':'away_R','FTHG':'away_RA'})
EPL2018_away_by_teams = EPL2018_away.groupby('Team')['away_games','away_win','away_R','away_RA'].sum().reset_index()
EPL2018_away_by_teams
```

## Step 8, 2017 Home + Away

```python
EPL2017_by_team = pd.merge(EPL2017_home_by_teams,EPL2017_away_by_teams, on='Team')

EPL2017_by_team['R'] = EPL2017_by_team['home_R'] + EPL2017_by_team['away_R']
EPL2017_by_team['RA'] = EPL2017_by_team['home_RA'] + EPL2017_by_team['away_RA']
EPL2017_by_team['pyth17'] = EPL2017_by_team['R']**2 / (EPL2017_by_team['R']**2 + EPL2017_by_team['RA']**2)

EPL2017_by_team['total_wins'] = EPL2017_by_team['home_win'] + EPL2017_by_team['away_win']
EPL2017_by_team['total_games'] = EPL2017_by_team['home_games'] + EPL2017_by_team['away_games']
EPL2017_by_team['wpc17'] = EPL2017_by_team['total_wins'] / EPL2017_by_team['total_games']

EPL2017_by_team
```

```python
EPL2017_by_team['diff'] = EPL2017_by_team['pyth17'] - EPL2017_by_team['wpc17']
EPL2017_by_team.sort_values(by='away_win')
```

## Step 9, 2018 Home + Away

```python
EPL2018_by_team = pd.merge(EPL2018_home_by_teams,EPL2018_away_by_teams, on='Team')
EPL2018_by_team['R'] = EPL2018_by_team['home_R'] + EPL2018_by_team['away_R']
EPL2018_by_team['RA'] = EPL2018_by_team['home_RA'] + EPL2018_by_team['away_RA']
EPL2018_by_team['pyth18'] = EPL2018_by_team['R']**2 / (EPL2018_by_team['R']**2 + EPL2018_by_team['RA']**2)

EPL2018_by_team['total_wins'] = EPL2018_by_team['home_win'] + EPL2018_by_team['away_win']
EPL2018_by_team['total_games'] = EPL2018_by_team['home_games'] + EPL2018_by_team['away_games']
EPL2018_by_team['wpc18'] = EPL2018_by_team['total_wins'] / EPL2018_by_team['total_games']

EPL2018_by_team
```

```python
EPL2018_by_team['diff'] = EPL2018_by_team['pyth18'] - EPL2018_by_team['wpc18']
EPL2018_by_team.sort_values(by='diff')
```

```python
EPL2018_by_team['diff_home_vs_away'] = EPL2018_by_team['home_win'] - EPL2018_by_team['away_win']
EPL2018_by_team.sort_values(by='diff_home_vs_away')
```

Merge 2017, 2018

```python
EPL_1718 = pd.merge(EPL2017_by_team,EPL2018_by_team, on='Team')

keyvars = EPL_1718[['Team','wpc17','wpc18','pyth17','pyth18']]
keyvars.corr()
```

## Part 3

Run the following regressions where the variable â€œscore_diffâ€ is the dependent variable.

Reg1_1: Include a single variable â€œhomeâ€ as the independent variable.

Reg1_2: Include the â€œhome,â€ â€œstadium_capacity,â€ â€œstadium_neutralâ€ variables and the interaction between â€œhomeâ€ and â€œstadium_neutralâ€ as the independent variables.

Reg1_3: Include the â€œhomeâ€, â€œstadium_capcity,â€ and â€œstadium_neutralâ€ variables, the interaction between â€œhomeâ€ and â€œstadium_neutral,â€ as well as â€œteamâ€ and â€œopponentâ€ as the independent variables.

```python
Reg1_1 =sm.ols(formula = 'score_diff ~ home', data= NFL_Game).fit()
Reg1_1.summary()
```

```python
NFL_Game["home_X_stadium_neutral"] = NFL_Game["home"]*NFL_Game["stadium_neutral"] 

Reg1_2 =sm.ols(formula = 'score_diff ~ home + stadium_capacity + stadium_neutral + home_X_stadium_neutral', data= NFL_Game).fit()
#or:
#Reg1_2 =sm.ols(formula = 'score_diff ~ home + stadium_capacity + stadium_neutral + home*stadium_neutral', data= NFL_Game).fit()

Reg1_2.summary()
```

```python
Reg1_3 =sm.ols(formula = 'score_diff ~ home + stadium_capacity + stadium_neutral + home_X_stadium_neutral + team + opponent', data= NFL_Game).fit()

Reg1_3.summary()
```

## Part 4

Run the following regressions where the variable â€œscoreâ€ is the dependent variable.

Reg2_1: Include â€œseasonâ€ and â€œhomeâ€ variables as independent variables.

Reg2_2: Include â€œseason,â€ â€œhome,â€ â€œweather_temperature,â€ â€œweather_wind_mph,â€ and â€œweather_humidityâ€ as independent variables.

Reg2_3: Include â€œseason,â€ â€œhome,â€ â€œweather_temperature,â€ â€œweather_wind_mph,â€ â€œweather_humidity,â€ â€œstadium_capacity,â€ â€œstadium_ageâ€, â€œstadium_type,â€ â€œstadium_neutralâ€, and the interaction between â€œhomeâ€ and â€œstadium_neutralâ€ as independent variables.     

Reg2_4: Include â€œseason,â€ â€œhome,â€ â€œweather_temperature,â€ â€œweather_wind_mph,â€ â€œweather_humidity,â€ â€œstadium_capacity,â€ â€œstadium_ageâ€, â€œstadium_type,â€ â€œstadium_neutralâ€, and the interaction between â€œhomeâ€ and â€œstadium_neutralâ€ as independent variables. Additionally, add both â€œteamâ€ and â€œopponentâ€ in the regression.

```python
Reg2_1 =sm.ols(formula = 'score ~ season + home', data= NFL_Game).fit()

Reg2_1.summary()
```

```python
Reg2_2 =sm.ols(formula = 'score ~ season + home + weather_temperature + weather_wind_mph + weather_humidity', data= NFL_Game).fit()

Reg2_2.summary()
```

```python
Reg2_3 =sm.ols(formula = 'score ~ season + home + weather_temperature + weather_wind_mph + weather_humidity \
               + stadium_capacity + stadium_age + stadium_type + stadium_neutral + home*stadium_neutral', data= NFL_Game).fit()

Reg2_3.summary()
```

```python
Reg2_4 =sm.ols(formula = 'score ~ season + home + weather_temperature + weather_wind_mph + weather_humidity \
               + stadium_capacity + stadium_age + stadium_type + stadium_neutral + home*stadium_neutral \
               + team + opponent', data= NFL_Game).fit()

Reg2_4.summary()
```
